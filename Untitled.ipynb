```python
from ctypes import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
mydll = cdll.LoadLibrary('pydll1.dll')
mydll.PrintTime.restype = c_char_p
print(str(mydll.PrintTime()))
```

```python
class SPID(Structure):
    
    __fields__ = [("kp", c_double),
                  ("ki", c_double),
                  ("kd", c_double),
                  ("set_point", c_double),
                  ("integration_sum", c_double),
                  ("p_term", c_double),
                  ("i_term", c_double),
                  ("d_term", c_double),
                  ("now_error", c_double),
                  ("last_error", c_double),]
    
    pass
```

```python
PIDControlPositionType = mydll.PIDControlPositionType
PIDControlPositionType.restype = c_double
PIDControlPositionType.argtypes = [POINTER(SPID)]
```

```python
pid = SPID()
```

```python
mydll.PIDConfigure(byref(pid), c_double(0.5), c_double(0.02), c_double(0.005), c_double(100))
```

```python
data = []
```
